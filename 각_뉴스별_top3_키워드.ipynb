```python
import pandas as pd

kbs = pd.read_json('/content/KBS.json', encoding='utf-8-sig')
kbs

kbs_comm = pd.read_json('/content/KBS_comment.json', encoding='utf-8-sig')
kbs_comm

keywords = pd.read_json('/content/news_keyword.json', encoding='utf-8-sig')
keywords
```

```python

kbs_news_and_comm = pd.merge(kbs, kbs_comm, how='left', left_on='nid', right_on = '번호')
kbs_keywords = pd.merge(kbs_news_and_comm, keywords, how='inner', on='nid')
kbs_keywords.sort_values('cnt', ascending=False).groupby('nid').head(3).sort_values('nid')
```
